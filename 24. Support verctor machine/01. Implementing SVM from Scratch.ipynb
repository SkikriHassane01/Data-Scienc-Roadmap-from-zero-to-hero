# Support Vector Machines (SVM) - A Comprehensive Guide

## 1. Introduction to SVM

### Linear Model

In Support Vector Machines, we aim to find the optimal hyperplane that separates two classes of data points with maximum margin. The margin is the distance between the closest data points of each class to the hyperplane.

### Decision Boundary Equations

The decision boundary is given by:

$$
w \cdot x_i - b \geq 1 \quad \text{if } y_i = 1
$$

$$
w \cdot x_i - b \leq -1 \quad \text{if } y_i = -1
$$

This can be combined into a single inequality:

$$
y_i (w \cdot x_i - b) \geq 1
$$

where \( y \in \{-1, 1\} \).

---

## 2. Hinge Loss Function

The SVM loss function, also known as the hinge loss, is used to penalize misclassified points or points that lie within the margin. The hinge loss function \( l \) is defined as:

$$
l = \max(0, 1 - y_i (w \cdot x_i - b))
$$

This can also be written as:

$$
l = 
\begin{cases}
0, & \text{if } y \cdot f(x) \geq 1 \\
1 - y \cdot f(x), & \text{otherwise}
\end{cases}
$$

---

## 3. Regularization Term and Cost Function

To avoid overfitting and encourage a larger margin, we add a regularization term to the hinge loss. The objective function \( J \) for SVM with regularization is:

$$
J = \lambda \| w \|^2 + \frac{1}{n} \sum_{i=1}^{n} \max(0, 1 - y_i (w \cdot x_i - b))
$$

Where:
- $( \lambda \| w \|^2 $) is the regularization term (to control the margin size),
- $( \max(0, 1 - y_i (w \cdot x_i - b)) $) is the hinge loss.

In cases:
- If $( y_i \cdot f(x) \geq 1 $), then $( J_i = \lambda \| w \|^2 $).
- Else, $( J_i = \lambda \| w \|^2 + 1 - y_i (w \cdot x_i - b) $).

---

## 4. Derivative of the Cost Function

To optimize the SVM objective, we need the gradient of \( J \) with respect to \( w \) and \( b \). The partial derivatives are:

For \( w \):
$$
\frac{\partial J}{\partial w} = 
\begin{cases} 
2 \lambda w & \text{if } y_i (w \cdot x_i - b) \geq 1 \\
2 \lambda w - y_i x_i & \text{otherwise} 
\end{cases}
$$

For \( b \):
$$
\frac{\partial J}{\partial b} = 
\begin{cases} 
0 & \text{if } y_i (w \cdot x_i - b) \geq 1 \\
-y_i & \text{otherwise} 
\end{cases}
$$

---

## 5. Update Rules

Using gradient descent, we update \( w \) and \( b \) as follows:

1. **For \( w \)**:
   $$ w = w - \eta \frac{\partial J}{\partial w} $$
   
2. **For \( b \)**:
   $$ b = b - \eta \frac{\partial J}{\partial b} $$

where \( \eta \) is the learning rate.

---

## 6. Steps to Train an SVM

1. **Initialize** \( w \) and \( b \) with small random values.
2. **Compute the hinge loss** and regularization term for each training example.
3. **Calculate the gradients** of the cost function with respect to \( w \) and \( b \).
4. **Update** \( w \) and \( b \) using gradient descent.
5. **Repeat** until convergence or for a fixed number of epochs.

---


```python
class SVM_scratch:
    
    def __init__(self, lr=0.001, lambda_param= 0.01, n_iters=1000):
        self.lr = lr
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None
    
    def compute_loss(self, X, y):
        """Calculate hinge loss."""
        y_ = np.where(y <= 0, -1, 1)
        distances = 1 - y_ * (np.dot(X, self.w) - self.b)
        losses = np.maximum(0, distances)  # Hinge loss
        hinge_loss = np.mean(losses) + self.lambda_param * np.sum(self.w ** 2)
        return hinge_loss    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        y_ = np.where(y <= 0, -1, 1)
        
        # initialize the weights and the bias 
        self.w = np.random.randn(n_features) * 0.01 # generate n_feature weight
        self.b = 0
        
        for i in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.lr * (2 * self.lambda_param * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.lr * y_[idx]

            # Print loss every 1000 iterations
            if i % 1000 == 0:
                loss = self.compute_loss(X, y)
                print(f"Iteration {i}, Loss: {loss:.4f}")
            
    def predict(self, X):
        approx = np.dot(X, self.w) - self.b
        return (approx >= 0).astype(int) # Return +1 or -1
    
    
    def draw_hyperplane(self, X, y):
        """Draw the decision boundary and margins for a 2D SVM."""
        plt.figure(figsize=(12, 6))
        sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='Set2')

        w = self.w
        b = self.b

        slope = -w[0] / w[1]  
        intercept = -b / w[1] 

        margin_distance = (1 / np.linalg.norm(w)) 

        # Generate x values for the line
        x_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)

        # Decision boundary (y = mx + c)
        y_vals = slope * x_vals + intercept

        # Margins (+1 and -1)
        y_vals_plus = slope * x_vals + intercept + margin_distance
        y_vals_minus = slope * x_vals + intercept - margin_distance

        # Plotting
        plt.plot(x_vals, y_vals, color='red', label='Decision Boundary')
        plt.plot(x_vals, y_vals_plus, color='blue', linestyle='--', label='+1 Margin')
        plt.plot(x_vals, y_vals_minus, color='green', linestyle='--', label='-1 Margin')

        plt.title("Scatter plot with decision boundary and margins")
        plt.legend()
        plt.show()
```

```python
svm = SVM_scratch(lr=0.01,n_iters=1000,lambda_param=0.001)
svm.fit(X_train, y_train)
```
