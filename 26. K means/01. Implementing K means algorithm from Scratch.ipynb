## How K-means Works: Step-by-Step

### Step 1: Choose the Number of Clusters (**K**)
- Select the number of clusters, **K**, that you want to divide your data into.
- **K** represents how many clusters you expect or want to identify in your dataset.

### Step 2: Initialize the Centroids
- Randomly select **K** points from your dataset to serve as the initial "centroids" (centers) of the clusters.

### Step 3: Assign Points to the Closest Centroid
- For each data point, calculate the distance between the data point and each centroid.
- Assign each data point to the cluster of the nearest centroid.

#### Distance Formula
The distance between a data point and a centroid is usually calculated using the Euclidean distance formula:

$$
\text{Distance} = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2 + \dots + (x_n - y_n)^2}
$$

where:
- $( x $) and $( y $) are the coordinates of the data point and the centroid.
- This formula computes the straight-line distance between each data point and the centroid.

### Step 4: Update the Centroids
- Once all points are assigned to clusters, update the centroids by calculating the mean (average) of all points in each cluster.
- Move each centroid to the new mean position. This makes the centroids more representative of the clusters.

### Step 5: Repeat
- Repeat Steps 3 and 4 until:
  - The centroids no longer change significantly, OR
  - A maximum number of iterations is reached.

At this point, the clusters are considered stable, and the algorithm stops.

## Choosing the Right Number of Clusters (**K**)
Choosing **K** can be challenging. The **Elbow Method** is commonly used to find an optimal **K**.

### Elbow Method
1. Run the K-means algorithm for various values of **K** (e.g., from 1 to 10).
2. For each **K**, calculate the sum of squared distances between data points and their cluster centroids.
3. Plot **K** versus the sum of squared distances. Look for an "elbow" point where the decrease in distance starts to slow down. This point is a good choice for **K**.

## Summary of Key Points

- **K-means is sensitive to the initial choice of centroids**. Different starting points can lead to different results.
- **K-means works best with well-separated, circular clusters**.
- **Data should be scaled properly** when features have different units or scales for better clustering results.

## Advantages and Limitations

**Advantages**:
- Simple to understand and implement.
- Fast and efficient for small to medium-sized datasets.

**Limitations**:
- Requires knowing **K** in advance.
- Doesnâ€™t work well with complex or irregular cluster shapes.
- Sensitive to outliers, which can distort the results.

```python
def euclidean_distance( X, Y):
    return np.sqrt(np.sum((X - Y) ** 2))

class Kmeans_scratch:
    
    def __init__(self, K=5, max_iters=100, plot_steps=False):
        assert K > 0, "K must be greater than 0"
        self.K = K
        self.max_iters = max_iters
        self.plot_steps = plot_steps
        
        # list of sample indices for each cluster
        self.clusters = [[] for _ in range(self.K)]
        
        # the centers (mean feature vector) for each cluster
        self.centroids = []
        
        
    def predict(self, X):
        self.X = X
        self.n_samples, self.n_features = X.shape

        # initialize the centroids
        random_sample_idxs = np.random.choice(self.n_samples, self.K, replace=False)
        self.centroids = [self.X[idx] for idx in random_sample_idxs]

        # Optimize clusters
        for _ in range(self.max_iters):
            # Assign samples to closest centroids (create clusters)
            self.clusters = self._create_clusters(self.centroids)

            if self.plot_steps:
                self.plot()

            # Calculate new centroids from the clusters
            centroids_old = self.centroids
            self.centroids = self._get_centroids(self.clusters)

            # check if clusters have changed
            if self._is_converged(centroids_old, self.centroids):
                break

            if self.plot_steps:
                self.plot()

        # Classify samples as the index of their clusters
        return self._get_cluster_labels(self.clusters)
    
    
    #### Helper functions #####
    #_________________________#
    
    ### 1. create clusters ###
    #________________________#

    def _create_clusters(self, centroids):
        # Assign the samples to the closest centroids to create clusters
        clusters = [[] for _ in range(self.K)]
        for idx, sample in enumerate(self.X):
            centroid_idx = self._closest_centroid(sample, centroids)
            clusters[centroid_idx].append(idx)
        return clusters
    
    def _closest_centroid(self, sample, centroids):
        # Distance of the current sample to each centroid
        distances = [euclidean_distance(sample, point) for point in centroids]
        closest_idx = np.argmin(distances)
        return closest_idx
    
    ### 2. get the new centroids ###
    #______________________________#
    
    def _get_centroids(self, clusters):
        # Assign mean value of clusters to centroids
        centroids = np.zeros((self.K, self.n_features))
        for cluster_idx, cluster in enumerate(clusters):
            cluster_mean = np.mean(self.X[cluster], axis=0)
            centroids[cluster_idx] = cluster_mean
            
        return centroids
    
    ### 3. check the convergence ###
    #______________________________#
    
    def _is_converged(self, centroids_old, centroids):
        # calculate the distance between old and new centroids, for all centroids
        distances = [euclidean_distance(centroids_old[i], centroids[i]) for i in range(self.K)]
        return sum(distances) == 0
    
    ### 4. get cluster label###
    #_________________________#
        
    def _get_cluster_labels(self, clusters):
        # each sample will get the label of the cluster it was assigned to
        labels = np.empty(self.n_samples)
        for cluster_idx, cluster in enumerate(clusters):
            for sample_idx in cluster:
                labels[sample_idx] = cluster_idx

        return labels
    

    def plot(self):
        fig, ax = plt.subplots(figsize=(12, 8))

        for i, index in enumerate(self.clusters):
            point = self.X[index].T
            ax.scatter(*point)

        for point in self.centroids:
            ax.scatter(*point, marker="x", color="black", linewidth=2)

        plt.show()
```
