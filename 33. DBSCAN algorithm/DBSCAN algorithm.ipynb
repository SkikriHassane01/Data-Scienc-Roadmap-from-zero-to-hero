# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

DBSCAN is a density-based clustering algorithm that groups together closely packed points, marking points in sparse regions as outliers. Unlike K-means, DBSCAN does not require the user to specify the number of clusters beforehand. It is particularly useful for discovering clusters with irregular shapes and for handling noise (outliers).

## Key Concepts of DBSCAN

- **Core Points**: Points that have a minimum number of points within a specified radius (epsilon, \( \epsilon \)).
- **Border Points**: Points that are within the \( \epsilon \)-radius of a core point, but do not have enough neighboring points to be a core point themselves.
- **Noise Points**: Points that are neither core nor border points; considered outliers.

## Parameters in DBSCAN

- **Epsilon (\( \epsilon \))**: The radius around a point to search for neighbors.
- **MinPts**: The minimum number of points required to form a dense region (core point).

The algorithm works by identifying the regions of high point density, separating them from regions of low point density.

---

## Steps in the DBSCAN Algorithm

1. **Choose Epsilon and MinPts**:
   - \( \epsilon \) is the neighborhood radius that defines how close points should be to each other.
   - MinPts is the minimum number of points needed to form a cluster.

2. **Find Core Points**:
   - For each point in the dataset, find its neighbors within \( \epsilon \).
   - If the number of neighbors is greater than or equal to MinPts, it becomes a core point.

3. **Cluster Expansion**:
   - For each core point, recursively add all points that are density-reachable (i.e., neighbors of neighbors).

4. **Label Border and Noise Points**:
   - Border points are labeled by the nearest core point, and points that are not reachable are labeled as noise.

---

## Advantages of DBSCAN

- Does not require specifying the number of clusters.
- Can detect clusters of arbitrary shape.
- Handles noise and outliers well.

## Disadvantages of DBSCAN

- Sensitive to the choice of $( \epsilon $) and MinPts.
- Struggles with clusters of varying densities.
- May have difficulties in high-dimensional spaces.


```python
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons


import warnings
warnings.filterwarnings("ignore")
```

```python
# Step 2: Create a Sample Dataset

# Generate moon-shaped data
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)

# Plot the data
plt.scatter(X[:, 0], X[:, 1], s=30, cmap='viridis')
plt.title("Generated Data (Moons)")
plt.show()
```

```python
# Step 3: Apply DBSCAN Algorithm


# Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', s=30)
plt.title("DBSCAN Clustering")
plt.show()
```

```python
# Step 4: Analyzing DBSCAN Results
# Display the labels
print("Labels from DBSCAN:", dbscan_labels[:10])

# Find the number of clusters (excluding noise points)
num_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
print(f'Number of clusters found: {num_clusters}')
```
