## first we will Standardize the data

Standardization is a preprocessing technique used in machine learning to rescale and transform the features (variables) of a dataset to have a mean of 0 and a standard deviation of 1.

$$
\text{Standardized value} = \frac{x - \mu}{\sigma}
$$

```python
def standardize_data(X_train, X_test):
    # calculate the mean and std
    
    mean = np.mean(X_train)
    std = np.std(X_train)
    
    X_train_st = (X_train - mean)/ std
    X_test_st = (X_test - mean)/ std
    
    return X_train_st, X_test_st
```

```python
X_train, X_test = standardize_data(X_train, X_test)
```

    4 - Model Implementation 🧠</h1>

Linear regression is used to predict continuous output values based on input features. The model function for linear regression can be expressed as:

$$
f_{w,b}(x) = wx + b
$$

where:
- $f_{w,b}(x)$: predicted output,
- $w$: weight parameter,
- $b$: bias parameter,
- $x$: input feature.

## **Training the Model**

Training a linear regression model involves finding the optimal values for $w$ and $b$ that best represent the relationship between the input and output in the dataset.

### 1. **Forward Pass (Prediction Step)**

In the forward pass, we compute the predicted output, $f_{w,b}(x)$, for each input $X$ using the current values of $w$ and $b$. This gives us the initial predictions of the model.

### 2. **Cost Function (Loss Measurement)**

The cost function evaluates the model's performance by quantifying the difference between predicted values and the actual values in the dataset. For linear regression, we use the Mean Squared Error (MSE) as the cost function:

$$
J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right)^2
$$

where:
- $J(w, b)$: cost (error) of the model,
- $m$: number of training examples,
- $x^{(i)}$: $i$-th input example,
- $y^{(i)}$: actual output for the $i$-th example.

The goal is to minimize $J(w, b)$ by adjusting $w$ and $b$.

### 3. **Backward Pass (Gradient Computation)**

The backward pass calculates gradients (partial derivatives) of the cost function with respect to $w$ and $b$. These gradients indicate how much to change $w$ and $b$ to reduce the error. The gradients are calculated as:

$$
\frac{\partial J(w,b)}{\partial b} = \frac{1}{m} \sum_{i=0}^{m-1} \left(f_{w,b}(X^{(i)}) - y^{(i)}\right)
$$

$$
\frac{\partial J(w,b)}{\partial w} = \frac{1}{m} \sum_{i=0}^{m-1} \left(f_{w,b}(X^{(i)}) - y^{(i)}\right) X^{(i)}
$$

### 4. **Parameter Update (Gradient Descent)**

Using gradient descent, we iteratively adjust $w$ and $b$ in the direction that reduces $J(w, b)$. The update equations are:

$$
w \leftarrow w - \alpha \frac{\partial J}{\partial w}
$$

$$
b \leftarrow b - \alpha \frac{\partial J}{\partial b}
$$

where:
- $\alpha$: learning rate, which determines the step size of each update.

Through this iterative process of computing predictions, measuring cost, calculating gradients, and updating parameters, the model learns the optimal values for $w$ and $b$, gradually improving its ability to fit the data accurately.

```python
class LinearRegression_scratch:
    """
    Parameters ==>:
    
    -   learning_rate(float) : the learning rate used in gradient descent
    -   convergence_tolerance(float, optional) : the tolerance for convergence (stopping criterion) = 1e-6
    
    Attributes ==>:
    
    - W(numpy.ndarray) : weights for linear regression model
    - b(float) : Intercept(bias) for our model
    
    Methods ==>:
    
    - __init__(lr, convergence_tol) => initialize model parameters
    
    - initialize_parameters(n_features): Initialize model parameters.
    
    - forward(X) => "X : our features" compute the formula of a line f(x) = a * X + b
    
    - Cost(prediction) => "prediction : output of the forward function" calculate the cost function
    
    - backward(prediction) => calculate the derivative 
    
    - fit(X, y, n_iter, plot=True) => fit the linear regression model to training data. 
    
    - predict(X) => "X: new data" predict the target value of a new data
    
    - save_model(filename=None) => save the training model into a pickle file 
    
    - load_model(filename) => load the training model using pickle
    """
    
    
    def __init__(self,lr, convergence_tol=1e-6):
        self.lr = lr
        self.convergence_tol = convergence_tol
        self.W = None
        self.b = None
        
        
    def initialize_parameters(self,n_features):
        """
        - initialize the weight with a random value from a standard normal distribution (mean:0, std:1) => then we will scale it down to get a small number to help model converge smoothly
        - initialize the intercept with 0
        """
        
        self.W = np.random.randn(n_features) * 0.01 # generate n_feature weight
        self.b = 0
    
    def forward(self,X):
        """"
        parameters :X (numpy.ndarray)
        return : prediction
        """
        
        return np.dot(X,self.W) + self.b
    
    
    def Cost(self, prediction, y):
        """
        parameters : 
        - prediction: the predicted value
        - y : the actual output 
        return: mean squared error
        """
        
        m = len(y)
        
        errors = (prediction - y) ** 2
        
        cost = (1/ (2 * m)) * np.sum(errors)
        
        return cost
    
    def backward(self, prediction, X, y):
        """
        parameters:
        - prediction: the predicted output 
        
        return:
        - the gradient of the weight
        - the gradient of the bias
        """
        
        m = len(self.W)
        
        self.dW = np.dot(prediction - y, X) / m
        self.db = (1/m) * (np.sum(prediction - y))
        
    
    def fit(self, X, y, n_iters, plot_cost=True):
        
        """
        X, y must be a numpy array
        X, y must have the same number of samples 
        n_iters must be greater then 0
        """
        
        assert isinstance(X, np.ndarray)
        assert isinstance(y, np.ndarray)
        assert X.shape[0] == y.shape[0]
        assert n_iters > 0
        
        self.X = X
        self.y = y
        n_features = X.shape[1]
        self.initialize_parameters(n_features)
        costs = []
        
        for i in range(n_iters):
            prediction = self.forward(X)
            
            cost = self.Cost(prediction, y)
            
            self.backward(prediction,X,y)
            
            self.W -= self.lr * self.dW
            
            self.b -= self.lr * self.db
            
            costs.append(cost)
            
            if i%100 == 0:
                print(f"Iteration: {i}, Cost: {cost}")
            
            if i>0 and abs(costs[-1] - costs[-2]) < self.convergence_tol:
                print(f'Congratulations! Converged after {i} iterations. with {costs[-1]} Cost')
                break
            
        if plot_cost:
            plt.figure(figsize=(12,6))
            plt.plot(range(len(costs)), costs, color = "blue")
            plt.title("Cost Vs Number of Iterations")
            plt.xlabel("Number of iteration")
            plt.ylabel("Cost")
            plt.grid(True)
            plt.show()
        
        
    def predict(self, X):
        """
        parameters: X: new testing data
        return: the prediction the y_output
        """
        
        return self.forward(X)
    
    def save_model(self, filename=None):
        """
        save the training model to a file using pickle
        """
        
        model = {
            'lr' : self.lr,
            'convergence_tol' : self.convergence_tol,
            'W' : self.W,
            'b' : self.b
        }
        
        with open(filename, 'wb') as f:
            pickle.dump(model,f)
        
    
    @classmethod
    def load_model(cls, filename):
        """
        Load a trained model from a file using pickle.
        
        Returns:
            LinearRegression: An instance of the LinearRegression class with loaded parameters.
        """
        with open(filename, 'rb') as file:
            model_data = pickle.load(file)

        # Create a new instance of the class and initialize it with the loaded parameters
        loaded_model = cls(model_data['learning_rate'], model_data['convergence_tol'])
        loaded_model.W = model_data['W']
        loaded_model.b = model_data['b']

        return loaded_model
        
```

```python
lr = LinearRegression_scratch(lr=0.00001)
lr.fit(X_train, y_train, n_iters=500, plot_cost=True)
```
