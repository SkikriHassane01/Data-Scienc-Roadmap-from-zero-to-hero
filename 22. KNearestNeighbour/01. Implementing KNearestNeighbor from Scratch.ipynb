# K-Nearest Neighbors (KNN) Algorithm Steps

## Step 1: Selecting the Optimal Value of K
- **K** represents the number of nearest neighbors to consider when making predictions.
- Choosing an optimal **K** is crucial, as it affects the accuracy of the model. Typically, **K** is chosen through cross-validation or by testing different values to find the best accuracy.

## Step 2: Calculating Distance
- To measure the similarity between the target and training data points, **Euclidean distance** is commonly used.
- Distance is calculated between each data point in the dataset and the target point using the formula:

  $$
  d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + \dots}
  $$

## Step 3: Finding Nearest Neighbors
- After calculating the distance, sort the data points by their distance to the target point.
- The **K** data points with the smallest distances to the target point are selected as the nearest neighbors.

## Step 4: Voting for Classification or Averaging for Regression
- **For Classification**: Perform majority voting on the **K-nearest neighbors** to determine the class label.
  - The class with the highest occurrence among the neighbors is chosen as the predicted class for the target point.
  
- **For Regression**: Calculate the average of the target values of the **K-nearest neighbors**.
  - The calculated average becomes the predicted output for the target data point.


```python
import numpy as np
from collections import Counter

class KNeighbors_scratch:
    """
    K-Nearest Neighbors (KNN) classification algorithm.

    Parameters:
    -----------
    n_neighbors : int, optional (default=5)
        Number of neighbors to use in the majority vote.

    Methods:
    --------
    euclidean_distance(point1, point2):
        Calculates the distance between two points.
    
    knn_predict(X, y, test_point, k):
        Finds the k-nearest neighbors and returns the predicted class.
    """
    
    def __init__(self, k=5):  
        self.k = k
    
    @staticmethod
    def euclidean_distance(point1, point2):
        """
        Calculates the Euclidean distance between two points.
        """
        return np.sqrt(np.sum((point1 - point2) ** 2))
    
    def knn_predict(self, X, y, test_point, k=None):
        """
        Predicts the class of the test_point based on the k-nearest neighbors.
        
        Parameters:
        -----------
        X : array-like, shape (n_samples, n_features)
            Training data.
            
        y : array-like, shape (n_samples,)
            Class labels for the training data.
            
        test_point : array-like, shape (n_features,)
            The point for which to predict the class label.
            
        k : int, optional
            The number of neighbors to consider. If not provided, defaults to self.k.
            
        Returns:
        --------
        int
            Predicted class label.
        """
        if k is None:
            k = self.k
        
        distances = []
        # Step 1: Calculate distances from test_point to all training points
        for i, point in enumerate(X):
            dist = self.euclidean_distance(test_point, point)
            distances.append((dist, y[i]))
        
        # Step 2: Sort distances and get the nearest k neighbors
        distances = sorted(distances, key=lambda x: x[0])
        k_nearest_neighbors = distances[:k]
        
        # Step 3: Count the classes of these neighbors
        classes = [neighbor[1] for neighbor in k_nearest_neighbors]
        most_common = Counter(classes).most_common(1)
        
        # Return the most common class among neighbors
        return most_common[0][0]
```
