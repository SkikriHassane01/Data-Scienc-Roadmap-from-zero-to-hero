    4 - How the algorithm works 🧠</h1>

1. starting with all examples at the root node then:
2. We'll calculate information gain for splitting on all possible features and pick the one with the highest value
3. Then we'll split the data according to the selected feature**
4. We'll repeat this  process until stopping criteria is met**

## Key Points:

### Gini Impurity

Gini Impurity is a measure used to quantify the "impurity" or "mixing" of classes within a dataset. It's commonly used in decision trees to determine the best attribute to split the data at each node. A lower Gini Impurity indicates a more "pure" subset of data (i.e., dominated by a single class), while a higher Gini Impurity shows a mixed class distribution.

### Formula
For a dataset with multiple classes, Gini Impurity $( G $) is calculated as:

$
G = 1 - \sum_{i=1}^{n} p_i^2
$

Where:
- $( p_i $) is the proportion of samples belonging to class $( i $),
- $( n $) is the number of unique classes.


### Entropy
**Entropy function which is a way to measure impurity**

**Entropy is represented by this function**
$$H = -\sum\limits_{}^{} p_{i}\text{log}_2 p_{i} 
$$

**Where $(p_1)$ is the fraction of examples that are a certain class**


### Information Gain

**Information gain is the reduction in entropy when he make a split**

**Recall that our goal is to choose the split that gives the highest information gain, information gain equation =**

**$$H(p_1^\text{node})- (w^{\text{left}}H(p_1^\text{left}) + w^{\text{right}}H(p_1^\text{right}))$$**
**where** 
- $H(p_1^\text{node})$ is entropy at the node 
- $H(p_1^\text{left})$ and $H(p_1^\text{right})$ are the entropies at the left and the right branches resulting from the split
- $w^{\text{left}}$ and $w^{\text{right}}$ are the proportion of examples at the left and right branch respectively

    5.1 - Question Class 🧠</h1>

```python
class Question:
    """
    attribute:
    ----------
    - column(int) : column number (eg 0 for color)
    - value : column value (eg Green for col 0)
    
    methods:
    --------
    match(example): Compare the feature value in an example (record of a dataset) to the
    feature value in this question.
    
    Returns True if it matches and False if it doesn't.
    """
    
    def __init__(self, num_col, value_col):
        self.num_col = num_col
        self.value_col = value_col
        
    def match(self, example):
        val = example[self.num_col] 
        if isinstance(val, (int, float)):  
            return val >= self.value_col
        else:
            return val == self.value_col
        
    def __repr__(self) -> str:
        condition = "=="
        if isinstance(self.value_col, (int, float)):  
            condition = ">="
        
        # whenever we create a question class this will print
        return "Is %s %s %s ?" % (
            col_name[self.num_col], condition, str(self.value_col)
        )
```

    5.2 - Partition the dataset 🧠</h1>

```python
def partition(train_data, question):
    """
    we will partition each train data  into two groups based on the question,
    for every row in the train data we will check if that row match the question, if so add this row to the
    true group, else add it to the false group.
    """
    
    true_group, false_group = [], []
    
    for row in train_data:
        if question.match(row):
            true_group.append(row)
        else:
            false_group.append(row)
    return true_group, false_group
```

```python
labels = ['A', 'A', 'B', 'B', 'B']
Counter(labels)
```

```python
def gini_impurity(labels):
    """
    Calculate the Gini impurity for a list of class labels.
    
    Parameters:
    ----------
    labels : list
        A list of class labels for a node (e.g., ['A', 'A', 'B', 'B', 'B'])
    
    Returns:
    -------
    float
        The Gini impurity for the node.
    """
    
    # Calculate the proportion of each class label
      # Count the frequency of each class in the labels
      
    label_counts = Counter(labels)
    total_samples = float(len(labels))
    
    impurity = 1.0
    for count in label_counts.values():
        prob = count / total_samples
        impurity -= prob ** 2
    return impurity    
    
```

    5.4 -  Information Gain</h1>

```python
def info_gain(left_labels, right_labels, current_impurity):
    """
    left : list of samples in the left node
    right : list of samples in the right node
    current_impurity : current impurity of the parent node
    """
    
    # Get only the class labels from left and right nodes

    proba = len(left_labels) / (len(left_labels) + len(right_labels))

    # Calculate and return the information gain
    return current_impurity - proba * gini_impurity(left_labels) - (1 - proba) * gini_impurity(right_labels)
```

    5.5 -  Find the best split</h1>

```python
def find_best_split(train_data):
    """
    find the best question to ask by iterating over every feature / value
    and calculating the information gain
    """
    
    best_gain = 0
    best_question = None
    labels = [lbl[-1] for lbl in train_data]
    current_impurity = gini_impurity(labels)
    n_feature = len(train_data[0]) - 1 # number of columns
    
    for col in range(n_feature):
        values = set([train[col] for train in train_data]) # get the unique value in each column
        
        for val in values:
            question = Question(col, val)
            
            # try splitting the dataset
            true_group, false_group = partition(train_data, question)
            
            # skip the split if it doesn't divide the dataset
            if len(true_group) == 0 or len(false_group) == 0:
                continue
            
            true_group = [true[-1] for true in true_group]
            false_group = [false[-1] for false in false_group]
            
            gain = info_gain(true_group, false_group, current_impurity)
            
            if gain >= best_gain:
                best_gain, best_question = gain, question

    return best_gain, best_question
            
```

    5.6 -  Build the Tree</h1>

```python
def class_counts(data):
    """Counts the number of each type of example in a dataset."""
    counts = {}  # a dictionary of label -> count.
    for row in data:
        label = row[-1]
        if label not in counts:
            counts[label] = 0
        counts[label] += 1
    return counts
```

```python
class_counts(train_data)
```

```python
class Leaf:
    """
    A leaf node : a node that has no children
    A leaf node classifies data.
    This holds a dictionary of class (e.g. "Apple") keys and the
    number of times it appears in the rows from the training data that reach this leaf.
    example :
    {
        "Apple": 3,
        "Orange": 2
    }

    """
    
    def __init__(self, train_data) -> None:
        self.predictions = class_counts(train_data)
    
```

```python
class Decision_Node:
    """
    A Decision Node holds a reference to the question, and the two child nodes
    """
    def __init__(self, question, true_branch, false_branch):
        self.question = question,
        self.true_branch = true_branch
        self.false_branch = false_branch
    
```

1. **Find the Best Split**: Identify the question (feature/value) that best separates the data.
2. **Check Base Case**: If no information gain is possible, create a leaf node with the final prediction.
3. **Partition Data**: Divide data into `true_rows` and `false_rows` based on the best question.
4. **Recursively Build**: Call `BuildTree` on each group (`true_branch` and `false_branch`) until reaching base cases.
5. **Create Decision Node**: Return a node that stores the question and connects the branches, building the tree.


```python
def BuildTree(train_data):
    """Decision Tree for classification, built using recursive splitting.

    Methods:
    --------
    build_tree(rows):
        Recursively constructs the decision tree based on information gain.
    """
    
    # Step 1: Find the best question to split the data
    gain, question = find_best_split(train_data)
    
    # Step 2: Base case - if no gain return a leaf node 
    if gain == 0:
        return Leaf(train_data)
    
    # Step 3: Partition data into two groups: true_rows and false_rows
    
    true_rows, false_rows = partition(train_data, question)
    
    # Step 4: Recursively build branches
    true_branch = BuildTree(true_rows)
    false_branch = BuildTree(false_rows)
    
     # Step 5: Return a Decision Node holding the best question and branches
     
    return Decision_Node(question, true_branch, false_branch)
```

```python
def PrintTree(node, spacing=""):
    """Prints the decision tree in a readable format."""
    
    # Base Case: If the node is a Leaf, print the predictions
    if isinstance(node, Leaf):
        print (spacing + "Predict", node.predictions)
        return

    question_str = str(node.question[0])

    print(spacing + question_str)

    # Recursive call to print the true branch
    print(spacing + '--> True:')
    PrintTree(node.true_branch, spacing + "  ")

    # Recursive call to print the false branch
    print(spacing + '--> False:')
    PrintTree(node.false_branch, spacing + "  ")
```

```python
def classify(train_data, node):
    """get the prediction of the data"""
    
    # base case 
    if isinstance(node, Leaf):
        return node.predictions
    if node.question[0].match(train_data):
        return classify(train_data, node.true_branch)
    else:
        return classify(train_data, node.false_branch)
```

```python
def print_leaf(counts):
    """A nicer way to print the predictions at a leaf."""
    total = sum(counts.values()) * 1.0
    probs = {}
    for lbl in counts.keys():
        probs[lbl] = str(int(counts[lbl] / total * 100)) + "%"
    return probs
```
